```python
#Take sentence input from the user
sentence = input("Enter sentence: ")

#Split the sentence into individual words using space as separator
words = sentence.split()

#Convert the list of words into a set to remove duplicate words
unique_words = set(words)

#Find the number of unique words using len()
count = len(unique_words)

#Print the results
print("Unique words count:", count)
print("Unique words:", unique_words)
```
